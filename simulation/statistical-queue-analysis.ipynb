```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Simulation parameters
arrival_rate = 8
service_rate = 10
hours = 50

queue = []
queue_length = 0

np.random.seed(42)

for _ in range(hours):
    arrivals = np.random.randint(arrival_rate-3, arrival_rate+3)
    services = np.random.randint(service_rate-2, service_rate+2)

    queue_length += arrivals
    queue_length -= min(queue_length, services)

    queue.append(queue_length)

# Visualization
plt.figure(figsize=(8,4))
plt.plot(queue)
plt.xlabel("Time (Simulation Hours)")
plt.ylabel("Queue Length")
plt.title("Simulated Queue Accumulation Behavior")
plt.grid(True)
plt.tight_layout()
plt.show()

df = pd.DataFrame({
    "time_index": range(len(queue)),
    "queue_remaining": queue
})

df.to_csv("simulation_output.csv", index=False, header=True)
```
